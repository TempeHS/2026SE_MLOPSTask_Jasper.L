### Feature Engineering

I'm using this notebook to apply data engineering processes to improve the performance of my machine learning model. I will be altering my features to imrpove how my model can use them.

```python
# Import frameworks
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data_frame = pd.read_csv("2.2.1.wrangled_data.csv")
```

#### Combine targets

I'm analysing the different targets, and deciding to combine certain ones based on my domain knowledge. This is to simplify the data for the model and reduce possible overfitting.

```python
data_frame["dominant_biome"].unique()
```

```python
data_frame["dominant_biome"] = data_frame["dominant_biome"].apply(
    lambda biome: "ocean" if "ocean" in biome else biome
)
data_frame["dominant_biome"] = data_frame["dominant_biome"].apply(
    lambda biome: "taiga" if "taiga" in biome else biome
)
data_frame["dominant_biome"] = data_frame["dominant_biome"].apply(
    lambda biome: "birch_forest" if "birch_forest" in biome else biome
)
data_frame["dominant_biome"].unique()
```

##### Encoding categorical variables

I am encoding my targets, changing them from strings to numerical values so that they are easier for my model to interact with and use.

```python
# Transforming biomes into numerical values
le = LabelEncoder()
le.fit(data_frame["dominant_biome"])

biome_mapping = dict(zip(le.classes_, range(len(le.classes_))))
print("Biome encoding map:")
for biome, number in biome_mapping.items():
    print(f"  {biome} -> {number}")

data_frame["dominant_biome"] = le.transform(data_frame["dominant_biome"])
print("\nEncoded column sample:")
print(data_frame["dominant_biome"].head())
```

#### Save the wrangled and engineered data to CSV

```python
data_frame.to_csv('../2.3.Model_Training/2.3.1.model_ready_data.csv', index=False)
```
