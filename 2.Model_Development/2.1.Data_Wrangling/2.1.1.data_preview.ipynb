#### Graphically present the data

At this early stage you are just wanting to evaluate the data, the below simple plots lets you look the data in different ways to inform your model design and data wrangling approaches.

(using multiple graphs to plot the most prominent features)
I have used multiple graph types to plot the most prominent features visually, allowing me to see outliers and trends between them.

```python
# Plot a histogram of a column
plt.hist(data_frame['minecraft:air'])
plt.title(f"Histogram of {data_frame['minecraft:air'].name}")
plt.ylabel('Count')
plt.xlabel(f'{data_frame["minecraft:air"].name}')
plt.show()
```

```python
# Scatter plot 2 columns to see the relationship
plt.scatter(data_frame['minecraft:air'], data_frame['dominant_biome'])
plt.title(f"Scatter of {data_frame['minecraft:air'].name} against {data_frame['dominant_biome'].name}")
plt.ylabel(f"{data_frame['dominant_biome'].name} Data")
plt.xlabel(f"{data_frame['minecraft:air'].name} Data")
plt.show()
```

```python
# Scatter plot multiples columns to see the relationship
x_plot = ['minecraft:air', 'minecraft:water']
for col in x_plot:
    plt.scatter(data_frame[col], data_frame['dominant_biome'], marker='x')
plt.title(f"Scatter of {*x_plot,} against {data_frame['dominant_biome'].name}")
plt.ylabel(f"{data_frame['dominant_biome'].name} Data")
plt.xlabel(f"{*x_plot,} Data")
plt.show()
```

```python
# Scatter plot 2 columns in separate charts with a shared y-axis
fig, (ax1, ax2) = plt.subplots(1,2, sharey=True)
plt.suptitle(f"Scatter of {data_frame['minecraft:air'].name} and {data_frame['minecraft:water'].name} against {data_frame['dominant_biome'].name}")
ax1.set_ylabel(f"{data_frame['dominant_biome'].name} Data")

ax1.scatter(data_frame['minecraft:air'], data_frame['dominant_biome'])
ax1.set_xlabel(f"{data_frame['minecraft:air'].name} Data")

ax2.scatter(data_frame['minecraft:water'], data_frame['dominant_biome'])
ax2.set_xlabel(f"{data_frame['minecraft:water'].name} Data")

plt.show()
```

```python
# 3D Scatter plot to compare water, air, and dirt_block

x_plot = ['minecraft:water', 'minecraft:dirt', 'minecraft:air']

fig = plt.figure()
plt.suptitle(f"3D Scatter of {x_plot[0]}, {x_plot[1]} against {x_plot[2]}")
ax = fig.add_subplot(111, projection='3d')

sc = ax.scatter(
    data_frame[x_plot[0]], data_frame[x_plot[1]], data_frame[x_plot[2]], 
    c=data_frame[x_plot[2]], cmap="viridis", s=10
)

ax.set_xlabel(x_plot[0])
ax.set_ylabel(x_plot[1])
ax.set_zlabel(x_plot[2])

plt.colorbar(sc, ax=ax, label=x_plot[2])
plt.show()
```
