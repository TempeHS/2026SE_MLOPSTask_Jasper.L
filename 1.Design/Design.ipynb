# Design

## Defining the business problem to be solved

Minecraft is an extremely popular survival/exploration game, where players must use the environment around them to their advantage, facing enemies and eventually beating a large final boss. I am creating an AI model to comlete minecraft. As an extra challenge it needs to infer information and not use the F3 menu, a menu that gives the player a lot of information about the current player and game state. It has the current capabilities to know what blocks and entities it is seeing and interact with the game, but I would like to give the model more knowledge of its surroundings, including knowing the biome it is currently in, which needs to be sensed through the blocks around it. I need to have a way to interact with the predicted biome from the output and assess its accuracy.

## Refactoring the business problem into a machine learning problem


I need to create a machine learning model to predict the prominent biome in a chunk based off of the blocks it contains.

## Defining success metrics



The model needs to predict the prominent biome in a chunk and output this information in a readable format, it needs to be accurate enough for another model to rely on, but can have occasional errors as chunks can contain more than 1 biome.
